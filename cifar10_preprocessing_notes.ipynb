```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
seed = 42
np.random.seed(seed=seed)
```

![image](res/global_contrast_norm.png)


```python
# creating an "image"
X = np.random.normal(size=(500, 500))

# scatter the "image"
plt.scatter(X[:,0], X[:,1])
plt.xlabel('x0')
plt.ylabel('x1')
plt.title('Raw input')
plt.show()
```

```python
mean = np.mean(X)
var = np.mean((X - mean)**2)
std = var**0.5

print(f'mean:                           {mean}')
print(f'variance:                       {var}')
print(f'standard deviation (contrast):  {std}')
```

```python
lamb = 0
eps = 1e-8
s = 1

X_new = s * ((X - mean) / max(eps, (lamb + var)**0.5))

# scatter the "image"
plt.scatter(X_new[:,0], X_new[:,1])
plt.xlabel('x0')
plt.ylabel('x1')
plt.title('Raw input')
plt.show()
```
